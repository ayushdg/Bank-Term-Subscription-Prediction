```python
import math

def pday(x):
    if x>10: return("More Than 10")
    elif x>0: return("1 to 10")
    elif x==0: return("Not Contacted")
    
def job(x):
    #if x=="unemployed": return("unemployed")
    #elif x in ["student","retired"]: return("non-working")
    #else: return("working")
    if x=="unemployed" : return("unemployed")
    elif x in ["student","retired"]: return("non-working")
    else: return("working")

#mappingEdu = {'basic.9y':'edu1', 'basic.6y':'edu1', 'basic.4y':'edu1',
#                 'high.school':'edu2', 'university.degree':'edu2', 'professional.course':'edu2',
#                 'unknown':'edu3'}
    
def marital(x):
    if x=="unknown": return("unknown")
    elif x in ["married","divorced"]: return("married-once")
    else: return("single")

#df[column+'_cat'] = pand.cut(df[column],cut_points,labels=categories)
#trainset = process_range(trainset,'age',[15,29,59,74,100],
#                            ["YoungAdult","Adult","Old Adult","Senior"])

def education(x):
    if x=="unknown": return("unknown")
    elif x in ["illiterate"]: return("illiterate")
    else: return("some-education")
    #elif (((x=="basic.9y") or (x=="basic.6y") or (x=="basic.4y")): return("edu_basic")
    #elif (x== "basic.9y" or "basic.6y" or "basic.4y"): return("edu_basic")
    #elif (x=="high.school" or "university.degree" or "professional.course"): return("edu_higher")
def enc(x):
    if x=="unknown": return(-1)
    elif x in ["yes"]: return(1)
    else: return(0)
    
def day(x):
    if x in ["mon","fri"]: return(0)
    else: return(1)
#duration contact and month not processed
training_data['label']=np.where(training_data['y']=='yes',1,0)
#training_data['newage']=[math.log(x) for x in training_data['age']]
training_data['new_duration']=[math.log(x+1) for x in training_data['duration']]
training_data['new_campaign']=[math.log(x) for x in training_data['campaign']]
#training_data['previous_flag']=np.where([(x>0) for x in training_data['previous']],1,0)
training_data['new_npdays']=np.where(training_data['pdays']==999,0,training_data['pdays'])
training_data['new_pdays']=[pday(x) for x in training_data['npdays']]
training_data['new_job']=[job(x) for x in training_data['job']]
training_data['new_marital']=[marital(x) for x in training_data['marital']]
training_data['new_education']=[education(x) for x in training_data['education']]
training_data['new_default']=[enc(x) for x in training_data['default']]
training_data['new_housing']=[enc(x) for x in training_data['housing']]
training_data['new_loan']=[enc(x) for x in training_data['loan']]
training_data['new_day']=[day(x) for x in training_data['day_of_week']]
training_data['poutcome_t']=np.where(training_data['poutcome']=='success',1,0)
training_data['leuribor3m']=[math.log(x) for x in training_data['euribor3m']]
training_data['newage']=pd.cut(training_data['age'],[15,29,59,74,100],labels=["YoungAdult","Adult","Old Adult","Senior"])
#          'previous',[-1,2,5,7],["Few","Lot","Much"]
training_data['previous_flag']=pd.cut(training_data['previous'],[-1,2,5,7],labels=["Few","Lot","Much"])

```

```python
column_names=['label', 'new_duration', 'new_campaign', 'new_npdays', 'new_pdays', 'new_job' , 'new_marital', 'new_education', 'new_default', 'new_housing', 'new_loan' , 'new_day', 'poutcome_t','leuribor3m','newage', 'previous_flag' ]

new_train = training_data[column_names]
```

```python
columns=new_train.columns.values.tolist()
numerical_columns=new_train._get_numeric_data().columns.values.tolist()
catergorical_columns=list(set(columns) - set(numerical_columns))

# encoding

dummies=pd.get_dummies(new_train[catergorical_columns], prefix=['col1', 'col2', 'col3', 'col4','col5', 'col6'])
train_data= pd.concat([new_train[numerical_columns], dummies], axis=1)
```

```python
#handling class imbalance
from sklearn.utils import resample
temp_train=train_data
value_1=temp_train[temp_train.label==1]
n=value_1.label.value_counts()
value_0=temp_train[temp_train.label==0]
#value_0.label.value_counts()
#value_0_sampled= resample(value_0,replace=False,n_samples=n,random_state=123)
value_0=temp_train[temp_train.label==0].sample(len(value_1.label))
temp_train_new=pd.concat([value_1,value_0])

X_train=temp_train_new.drop(["label"],axis=1)
y_train=temp_train_new['label']
```

```python
from sklearn.model_selection import KFold

from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
model=RandomForestClassifier() 
errors=[]
acc=[]
for train_index, test_index in KFold(n_splits=10).split(X_train1):
    X_train2, X_test2 = X_train1.iloc[train_index], X_train1.iloc[test_index]
    Y_train2, Y_test2 = y_train.iloc[train_index], y_train.iloc[test_index]

    lm = model.fit(X_train2,Y_train2)
    Y_pred = lm.predict(X_test2)
    test_accu=accuracy_score(Y_test2,Y_pred)
    acc.append(test_accu)
    errors.append(mean_absolute_error(Y_test2, Y_pred))

print("mae is" , np.mean(errors))
print("mae is" , np.mean(acc))
```
